#`What is Data Frame`

A DataFrame is a data structure that organizes data into a 2-dimensional table of rows and columns, much like a spreadsheet. DataFrames are one of the most common data structures used in modern data analytics because they are a flexible and intuitive way of storing and working with data.

#Uzbek

#`Data Frame nima`

DataFrame - bu elektron jadval kabi ma'lumotlarni ikki o'lchovli satrlar va ustunlar jadvaliga joylashtiradigan ma'lumotlar strukturasi. DataFrames zamonaviy ma'lumotlar tahlilida qo'llaniladigan eng keng tarqalgan ma'lumotlar tuzilmalaridan biridir, chunki ular ma'lumotlarni saqlash va ular bilan ishlashning moslashuvchan va intuitiv usulidir.

```python
import pandas as pd
```

```python
data = {
    'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
    'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
    'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]
}

df = pd.DataFrame(data)
```

```python
df # Malumotlar
```

```python
df.head() # Bu method orqali ma'lumotlarimizniboshidagi 5 tasini chiqaramiz
```

```python
df.head(3) # Bu usul orqali o'zimiz hohlagan ma'lumotlarimiz ichidagi elementlarni chiqarishimiz mumkun
```

```python
df.shape # Bu method orqali ma'lumotimiz o'chamini ko'rishimiz mumkun
```

```python
df = pd.DataFrame(data, columns=['Yil','Temp', 'Aholi soni', 'YAIM']) # Bu method orqali ustunlar qatorini o'zgartirishimiz mumkun 
# columns meaning `ustun`
```

```python
df
```

```python
df = pd.DataFrame(data, columns=['Yil','Temp', 'Aholi soni', 'YAIM'], index = [1, 2, 3, 4, 5, 6, 7, 8])
```

```python
df
```

```python
aholi = df['Aholi soni'] # Bu method orqali har bir qatorni ajratib olishimiz mumkun Series ma'lumotlar turiga
```
